```python
#!/usr/bin/env python3
"""
scripts/compile_global_ebola_archive.py
Dynamic Ebola Pipeline Engine - Advanced Variable Mapping Edition
Runs via relative paths from the script directory.
"""

import os
import csv
import json
import urllib.request
from datetime import datetime

# Define relative directory pathing based on your repository architecture
# Executed from: /Users/koissi/Desktop/Ebola/Ebola_dashboard/script
# Output Target:  /Users/koissi/Desktop/Ebola/Ebola_dashboard/data/ebola_comprehensive.csv
SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
OUTPUT_PATH = os.path.normpath(os.path.join(SCRIPT_DIR, "..", "data", "ebola_comprehensive.csv"))

# CDC/WHO Complete Tracking Data Source Mirror (Combines Suspected, Confirmed, Deaths, Recoveries)
CDC_WHO_MIRROR = "https://raw.githubusercontent.com/datasets/epidemic-ebola/master/data/ebola-cases-unhcr.csv"

def normalize_country(c_str):
    """Standardizes country strings to a clean uniform format."""
    if not c_str:
        return "Unknown"
    c_lower = c_str.lower()
    if "congo" in c_lower or "rdc" in c_lower or "zaire" in c_lower:
        return "DR Congo"
    if "sierra" in c_lower:
        return "Sierra Leone"
    if "states" in c_lower:
        return "United States"
    return c_str.strip().title()

print("=====================================================================")
print("RUNNING VARIABLE-COMPLETE EBOLA EXTRACTOR PIPELINE")
print("=====================================================================")
print(f"-> Source Execution Context: {SCRIPT_DIR}")
print(f"-> Target Relative Destination: {OUTPUT_PATH}")

master_pool = []

try:
    print("-> Connecting to CDC/WHO Historical Data API Endpoint Mirror...")
    req = urllib.request.Request(
        CDC_WHO_MIRROR, 
        headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)'}
    )
    
    with urllib.request.urlopen(req, timeout=25) as response:
        raw_csv = response.read().decode('utf-8')
        lines = raw_csv.strip().split('\n')
        reader = csv.DictReader(lines)
        
        print("   ✔ Connected successfully. Ingesting rows and harvesting metrics...")
        
        for row in reader:
            # Map dynamic schema keys from the source (Date, Country, Variant, Value)
            date_raw = row.get('Date') or row.get('date')
            country_raw = row.get('Country') or row.get('country')
            metric_type = row.get('Variant') or row.get('variant') or ''
            val_str = row.get('Value') or row.get('value') or '0'
            
            if not date_raw or not country_raw:
                continue
                
            try:
                val_numeric = float(val_str)
            except ValueError:
                continue
            
            country_clean = normalize_country(country_raw)
            
            # Dynamic Classification Routing Layer (Eliminating 'NA' bias)
            # Initializes tracking metrics dynamically based on row observations
            susp = "NA"
            conf = 0.0
            dead = 0.0
            recov = "NA"
            
            m_lower = metric_type.lower()
            if "suspected" in m_lower or "suspect" in m_lower:
                susp = val_numeric
            elif "confirmed" in m_lower or "confirme" in m_lower:
                conf = val_numeric
            elif "death" in m_lower or "deces" in m_lower:
                dead = val_numeric
            elif "recovered" in m_lower or "gueri" in m_lower or "recovery" in m_lower:
                recov = val_numeric
            else:
                # Fallback assignment for general baseline metrics
                conf = val_numeric 

            master_pool.append({
                "date": date_raw.strip(),
                "country": country_clean,
                "suspected_cases": susp,
                "confirmed_cases": conf,
                "deaths": dead,
                "recoveries": recov,
                "source_url": CDC_WHO_MIRROR
            })

except Exception as e:
    print(f"❌ Network transaction mapping error: {e}")

# --- PROCESSING & RELATIVE STORAGE EXECUTION ---
if master_pool:
    print("-> Aggregation complete. Running optimization and deduplication passes...")
    
    final_dataset = []
    seen_keys = set()
    
    # Sort data backwards chronologically
    for record in sorted(master_pool, key=lambda x: x['date'], reverse=True):
        conf_val = record["confirmed_cases"]
        dead_val = record["deaths"]
        
        # Calculate case fatality rate dynamically using actual data points
        record["case_fatality_rate_pct"] = round((dead_val / conf_val) * 100, 2) if conf_val > 0 else 0.0
        
        # Composite primary key boundaries evaluation to prevent row doubling
        composite_fingerprint = (record['date'], record['country'], record['suspected_cases'], conf_val, dead_val, record['recoveries'])
        if composite_fingerprint not in seen_keys:
            seen_keys.add(composite_fingerprint)
            final_dataset.append(record)
            
    # Commit changes out to disk relative storage path
    os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
    target_fields = ["date", "country", "suspected_cases", "confirmed_cases", "deaths", "recoveries", "case_fatality_rate_pct", "source_url"]
    
    with open(OUTPUT_PATH, 'w', newline='', encoding='utf-8') as target_file:
        writer = csv.DictWriter(target_file, fieldnames=target_fields)
        writer.writeheader()
        writer.writerows(final_dataset)
        
    print("=====================================================================")
    print(f"✔ PIPELINE SYNC COMPLETE!")
    print(f"✔ File Saved to: {OUTPUT_PATH}")
    print(f"✔ Extracted distinct row vectors: {len(final_dataset)}")
    print("=====================================================================")
else:
    print("❌ Critical Execution Failure: No data frames could be compiled.")
    print("=====================================================================")
```
